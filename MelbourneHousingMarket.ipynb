```python
# Her sütundaki eksik değerlerin sayısını hesaplayın
df.isnull().sum()
```

```python
# Seaborn kütüphanesini içeri aktarın
import seaborn as sns
```

```python
# Seaborn temasını ayarlayın ve çizimlerin boyutunu ve çözünürlüğünü belirleyin
sns.set_theme()
sns.set(rc={"figure.dpi":300,"figure.figsize":(12,9)})
```

```python
# Eksik değerleri görselleştirmek için ısı haritasını kullanın
sns.heatmap(df.isnull(), cbar=False)
```

```python
# "Price" sütununun ortanca değerini hesaplayın
price_median = df["Price"].median()
price_median
```

```python
# Eksik değerleri "Price" sütununun ortanca değeri ile doldurun
# fillna() yöntemi eksik değerleri belirli bir değerle doldurmanıza olanak tanır.
# Bu durumda, "Price" sütunundaki eksik değerler "price_median" ile doldurulur.
# inplace=True parametresi, değişikliğin doğrudan veri çerçevesine uygulanmasını sağlar,
# yani yeni bir değişken oluşturulmaz.
df["Price"].fillna(price_median, inplace=True)
```

```python
# "Price" sütununda hala eksik değerlerin olup olmadığını kontrol edin
df["Price"].isnull().sum()
```

```python
# "Price" sütununun istatistiksel özetini görüntüleyin
df["Price"].describe().round()
```

```python
# Her sütundaki eksik değerlerin sayısını tekrar kontrol edin
df.isnull().sum()
```
